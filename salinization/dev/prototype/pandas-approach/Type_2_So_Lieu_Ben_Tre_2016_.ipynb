* Code column should be called 'code'
* Station code is one word, all upper case: BINHDAI, ANTHUAN, BENTRAI, LOCTHUAN, SONDOC, ANTHUAN, etc.
* Date column should be called 'date'
* Date should contain only the date part in yyyy-MM-dd format
** 'time' column is missing: fill blank if you don't have it
* 'min' and 'max' columns are right after 'time' column
* Column '1' ... '23' should be renamed to 'h01' ... 'h24'
* output file should be in .CSV format and named as yyyy-stationcode.csv

```python
station =df.iloc[2][0]
#stations = df.columns[0]
station = station[station.find(':') + 1:].strip()
station = unidecode.unidecode(station).upper().replace(' ', '')
print(station)
```
